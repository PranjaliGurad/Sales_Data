```python
x = df1.drop(["Item_Outlet_Sales"],axis = 1)
y = df1["Item_Outlet_Sales"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.8 , random_state=0)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# Model Evaluation for Training Data

y_pred_train = model.predict(x_train)

mse = mean_squared_error(y_train, y_pred_train)
print("Mean Sqaured Error :",mse)

rmse = np.sqrt(mse)
print("Root Mean Sqaured Error :",rmse)

mae = mean_absolute_error(y_train, y_pred_train)
print("Mean Absolute Error :",mae)

r_squared_value = r2_score(y_train, y_pred_train)
print("R Squared Value :",r_squared_value)
```

```python
# Model Evaluation for Testing Data
y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
print("Mean Sqaured Error :",mse)

rmse = np.sqrt(mse)
print("Root Mean Sqaured Error :",rmse)

mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error :",mae)

r_squared_value = r2_score(y_test, y_pred)
print("R Squared Value :",r_squared_value)
```

## 2 no multicolinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df2 = df1.drop(["Item_Outlet_Sales"], axis = 1)
df3 = df1["Item_Outlet_Sales"]
df2 = pd.concat([df2,df3], axis = 1)
df2
x = df2.iloc[:,:-1]
y = df2.iloc[:,-1]
```

```python
vif_list = []
for i in range(x.shape[1]):
    vif = variance_inflation_factor(x.to_numpy(),i)
    vif_list.append(vif)
    
s = pd.Series(vif_list, index= x.columns)    
```

```python
s.sort_values().plot(kind = "barh",figsize = (3,10))
```
